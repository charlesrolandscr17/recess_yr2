### 5. Write a Pandas program to display the most frequent value in the given series and replace everything else as ‘Other’ in the series. (Use the ‘HomeTeamGoals’ column in the dataset provided) 

```python
df = pd.read_csv("AfricaCupofNationsMatches.csv")
home_team_goals = df["HomeTeamGoals"]
home_team_goals.describe()
```

```python
def not_frequent(val):
    if home_team_goals.mode()[0] != val:
        return "Other"
    return val
home_team_goals.apply(not_frequent)
```
